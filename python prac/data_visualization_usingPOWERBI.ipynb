```python
import numpy as np 
import pandas as pd

import matplotlib.pyplot as plt 
df=pd.read_csv('datafile (2).csv')
 
# set width of bar 

barWidth = 0.25

fig = plt.subplots(figsize =(12, 8)) 

 
# set height of bar 

production06_07=df['Production 2006-07']
crop=df['Crop             ']
production07_08=df['Production 2007-08']
production08_09=df['Production 2008-09']


 
# Set position of bar on X axis 

br1 = np.arange(len(crop)) 

br2 = [x + barWidth for x in br1] 

br3 = [x + barWidth for x in br2] 
 
# Make the plot

plt.bar(br1, production06_07, color ='r', width = barWidth, 

        edgecolor ='grey', label ='Production 2006-07') 

plt.bar(br2, production07_08, color ='g', width = barWidth, 

        edgecolor ='grey', label ='Production 2007-08') 

plt.bar(br3, production08_09, color ='b', width = barWidth, 

        edgecolor ='grey', label ='Production 2008-09') 
 
# Adding Xticks 

plt.xlabel('Crops', fontweight ='bold', fontsize = 15) 



plt.xticks([r + barWidth for r in range(len(crop))], 

        ['Total Foodgrains','Rice', 'Wheat', 'Jowar', 'Bajra', 'Maize','Ragi','Small millets','Barley','Cereals','Arhar','Onion','Sugarcane','Tobacco'])
 
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_csv('datafile (2).csv')
 
# use to set style of background of plot
sns.set(style="whitegrid")
 
# loading data-set
production06_07=df['Production 2006-07']
crop=df['Crop             ']
production07_08=df['Production 2007-08']
production08_09=df['Production 2008-09']


 
# plotting strip plot with seaborn
# deciding the attributes of dataset on
# which plot should be made
sns.barplot(crop,production07_08 )
 
# giving title to the plot
plt.title('Graph')
 
# function to show plot
plt.show()
```
